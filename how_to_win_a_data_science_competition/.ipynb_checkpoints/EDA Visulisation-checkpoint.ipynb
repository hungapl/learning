```python
from sklearn.datasets import load_wine
import pandas as pd
from matplotlib import pyplot as plot
%matplotlib inline

def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df

data = sklearn_to_df(load_wine())

feature_cols = ['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
       'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
       'proanthocyanins', 'color_intensity', 'hue',
       'od280/od315_of_diluted_wines', 'proline']
X = data[feature_cols]
Y = data.target.values
```

```python
# Plot pairs of features
pd.plotting.scatter_matrix(X,c = Y, figsize=(15, 15), marker='o',hist_kwds={'bins': 20}, s=60, alpha=.8);
```

```python
# Plot correlation matrix
corr = data.corr()
#lt.matshow();
#lt.xticks(range(len(X.columns)), X.columns);
#lt.yticks(range(len(X.columns)), X.columns);
corr.style.background_gradient(cmap='coolwarm')
```

```python
# Index versus value plot
plt.scatter(range(len(X)), data.sort_values(by='target')['od280/od315_of_diluted_wines'].values, c=Y, marker='o');
```

```python
data.mean().sort_values().plot(style='.');
```
